```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Heights=[70,72,75,78,83,72,72,84]
weights=[180,190,200,240,218,200,172,170]

plt.scatter(Heights,weights,label='Heights vs Weights')
plt.xlim(0,100)
plt.ylim(0,300)

plt.xlabel("Height")
plt.ylabel("Weight")

plt.grid()
plt.legend()
plt.show()
```
