```python
import matplotlib.pyplot as plt
import numpy as np
X=np.array(['Green','Blue','Red','Purple','Pink','Orange'])
y=np.array([8,14,12,5,8,2])

plt.bar(X,y,color='orange')
plt.xlabel("Color")
plt.ylabel("Frequency")

plt.legend("Survey Results: What is your fav colors?")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics

X=np.array([ 20000,25000,30000,35000,40000,45000,50000,55000,60000,
    65000,70000,75000,80000,85000,90000,95000,100000,105000,110000
])
y=np.array([100,80,84,145,150,160,180,200,225,200,190,195,180,165,140,140,80,65,30])

plt.bar(X, y, width=4000,color='blue', edgecolor='black', alpha=0.7)

mean=statistics.mean(y)
plt.axhline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')

median=statistics.median(y)
plt.axhline(median, color='green', linestyle='--', label=f'Median: {median:.2f}')

mode=statistics.mode(y)
plt.axhline(mode, color='yellow', linestyle='--', label=f'mode: {mode:.2f}')

for i in range(len(X)):
    plt.text(X[i], y[i], str(y[i]), ha="center",color='black')

plt.xlabel("Salary Range")
plt.ylabel("Frequency")
plt.title("Salary Distribution")
plt.xticks(rotation=45)

plt.tight_layout()
plt.legend()
plt.show()
```
