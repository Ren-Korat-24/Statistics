A boxplot (sometimes called a box-and-whisker plot) is a plot that shows 
the five-number summary of a dataset. 
The five-number summary include: 
1.The minimum 
2.The first quartile 
3.The median 
4.The third quartile 
5.The maximum 
A boxplot allows us to easily visualize the distribution of values in a dataset 
using one simple plot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
#Boxplot
height = [14,16,12,11,24,19,13,12,20,10]

plt.boxplot(height,vert=False, patch_artist=True,
            boxprops={'facecolor': 'lightblue', 'edgecolor': 'black'}, medianprops={'color': 'red'},
            whiskerprops={'color': 'green'})

min=np.min(height)
plt.text(min,1.1,f"Min={min}",ha='center')

q1=np.percentile(height,25)
plt.text(q1,1.1,f"Q1={q1}",ha='center')

median=np.median(height)
plt.text(median,1.1,f"Median={median}",ha='left')

max=np.max(height)
plt.text(max,1.1,f"Max={max}",ha='center')

q3=np.percentile(height,75)
plt.text(q3,1.1,f"Q3={q3}",ha='center')

plt.title('Boxplot of Heights(inches)')
plt.xlabel('Height(inches)')

plt.show()
```
