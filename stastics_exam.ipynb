```python
#Box Plot
import numpy as np
import matplotlib.pyplot as plt
weights=[37 ,40, 39, 40.5, 42, 51, 41.5 ,39 ,41 ,30, 40 ,42 ,40.5 ,39.5 ,41 ,40.5 ,37 ,39.5 ,40 ,41 ,38.5, 39.5, 40, 41, 39 ,40.5 ,40 ,38.5 ,39.5 ,41.5]

plt.boxplot(weights,vert=False)

min=np.min(weights)
plt.text(min,1.1,f"Min={min}",ha='center')

q1=np.percentile(weights,25)
plt.text(q1,1.1,f"Q1={q1}",ha='right')

median=np.median(weights)
plt.text(median,1.1,f"Median={median}",ha='center')

max=np.max(weights)
plt.text(max,1.1,f"Max={max}",ha='center')

q3=np.percentile(weights,75)
plt.text(q3,1.1,f"Q3={q3}",ha='left')

plt.title(" The weight of those pupils was measured in kg")
plt.xlabel("Weights")
plt.show()
```
